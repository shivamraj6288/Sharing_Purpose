```python
import numpy as np
from matplotlib.pylab import plt
def projectile(angle, v0, time, cd=0):
    t = np.linspace(0,time,300)
    dt = t[1] - t[0] #s
    g = 9.8 # m/s2
    m = 1 # kg
    # Distribute initial velocity to x and y.
    v0 = v0 # m/s
    vx = v0*np.cos(angle)
    vy = v0*np.sin(angle)
    y0 = 0
    x0 = 0
    vg = 0
    vy_list=[vy]
    vx_list=[vx]
    y_list = [y0]
    x_list = [x0]
    ax=0
    ay= -g
    acdy=0;
    acdx=0;
    for i in range(1, len(t)):
        # Update y.
        fd=cd*v0**2
#         print ("cd value :-", cd)
        tan=abs(vy/vx);
        fdy=fd*np.sin(np.arctan(tan))
        fdx=fd*np.cos(np.arctan(tan))
        acdx=-fdx;
        if (vy>0):
            acdy=-fdy*dt
        else:
            acdy=fdy*dt
        vg = ay * dt
        vy=vy+vg+acdy
        y0 += vy* dt
        # Update x.
        if (vx>0):
            vx=vx+(acdx*dt)
        else :
            vx=0
        x0 += vx * dt
        # Stop updating when it hits ground at y=0.
        if y0 > 0:
            y_list.append(y0)
            x_list.append(x0)
            vy_list.append(vy)
            vx_list.append(vx)
            v0=(vx**2+vy**2)**(0.5)
        else:
            break
    # (max x, max t)
    return x_list, y_list,vx_list,vy_list
```

```python
x_list1,y_list1,vx_list,vy_list=projectile(np.pi/3,100,30)
# print (x_list1,y_list1)


```

```python
plot1=plt.figure(1)
plt.plot(x_list1,y_list1)
plt.xlabel("x")
plt.ylabel("y")
x_list2,y_list2,vx_list2,vy_list2=projectile(np.pi/4,60,25)
plt.plot(x_list2,y_list2)
plt.title("Combine graph of two different projectiles without drag force")
plt.show()
range2=max(x_list2)
print (range2)
```
